```python
import google.generativeai as genai
```

```python
genai.configure(api_key="MY-ACTUAL-KEY")
```

```python
model = genai.GenerativeModel("gemini-3-flash-preview")
```

```python
prompt = "A train travels 60 km in 1 hour 30 minutes. What is its average speed?"
print(prompt)
```

```python
response = model.generate_content(
    prompt,
    generation_config={"temperature": 0}
)

print("Basic Output:\n")
print(response.text)
```

```python
cot_prompt = prompt + " Think step by step."

response = model.generate_content(
    cot_prompt,
    generation_config={"temperature": 0}
)

print("Chain-of-Thought Output:\n")
print(response.text)
```

```python
print("Self-Consistency Outputs:\n")

for i in range(5):
    response = model.generate_content(
        cot_prompt,
        generation_config={"temperature": 0.7}
    )
    print(f"Run {i+1}:")
    print(response.text)
    print("-" * 40)
```

```python
# Temperature = 0
response_low = model.generate_content(
    cot_prompt,
    generation_config={"temperature": 0}
)

print("Temperature = 0:\n")
print(response_low.text)

print("\n" + "="*60 + "\n")

# Temperature = 0.7
response_high = model.generate_content(
    cot_prompt,
    generation_config={"temperature": 0.7}
)

print("Temperature = 0.7:\n")
print(response_high.text)
```
